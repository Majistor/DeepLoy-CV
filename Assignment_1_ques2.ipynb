```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(name , i , Title):
    plt.subplot( 5, 1 ,i)
    plt.imshow(name, cmap='gray')
    plt.title(Title)
    plt.axis('off')
```

```python
cap = cv2.VideoCapture(0)
ret, frame1 = cap.read()
cap.release()

frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
```

```python
cap = cv2.VideoCapture(0)
ret2, frame2 = cap.read()
cap.release()

frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
```

```python
# Apply a low-pass filter (Gaussian Blur) to frame2
low_pass = cv2.GaussianBlur(frame2, (25, 25), 0)
```

```python
# Apply a high-pass filter to frame1
low_freq_img1 = cv2.GaussianBlur(frame1, (25, 25), 0)
high_pass = cv2.subtract(frame1, low_freq_img1)
```

```python
# Combine high-pass and low-pass images
hybrid_frame = cv2.add(high_pass, low_pass)
```

```python
titles = ['Original Image 1',  'Original Image 2', 'High-Pass Filtered','Low-Pass Filtered', 'Hybrid Image']
images = [frame1,frame2, high_pass, low_pass, hybrid_frame]
```

```python
for i in range(5):
    show(images[i], i+1, titles[i])
```
